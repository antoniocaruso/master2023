Quindi il guadagno tra l'acquisto in data A, e la vendita in data B, sarà uguale al valore
di 
$$ \sum_{i=A}^{B} l_i $$
ed il problema di massimizzazione del guadagno si risolve, massimizzando l'espressione sopra, cioè, cercando i valori di A,B tali che:
$$ max_{A,B}  \sum_{i=A}^{B} l_i  $$

cercando la sottosequenza [A,B] del vettore l, tale che la somma sia massima.


```python
#primo tentativo di soluzione, usiamo letteralmente la definizione del problema.

def solve_one(v):
    """ solve the maximum sum subsequence problem, solution 1 """
    # iterate all possibile subsequence with B > A
    # find the maximum
    max_sum = -10000000
    A,B = 0,0                # solutions
    n = len(v)
    for i in range(n):
        for j in range(i,n):
            s = sum(v[i:j])
            if s > max_sum:
                max_sum = s
                A,B = i,j
    return (A,B,max_sum)
```

Ogni funzione/problema ha un insieme di valori di input, sia $D$ tale insieme.
Per ogni elemento $d \in D$, definiamo $t(d)$ il tempo impiegato dall'algoritmo
per terminare con un risultato (in passi di esecuzione).

Definiamo sui valori $d \in D$ una misura di 'lunghezza' o 'dimensione', quindi, per ogni $d \in D$ chiamiamo $n = |d|$ la lunghezza o dimensione dell'istanza $d$. Definiamo *Complessità Computazionale* di un Algoritmo la funzione:

$$T(n) = \max_{\forall d \in D, |d| = n} t(d)$$

Notiamo che intuitivamente, $T(n)$ è una funzione crescente, o comunque non-decrescente, inoltre sulle istanze della stessa dimensione quello che conta sarà il tempo impegnato dall'Algoritmo per risolvere la peggiore possibile (il massimo). Questo modo di definire la funzione di complessità, viene detto appunto di *caso peggiore*, e serve per garantire un limite superiore entro cui è garantito che il nostro algoritmo terminerà l'esecuzione.
La complessità del primo Algoritmo non è delle migliori. Infatti, su una lista di lunghezza $n$, le istruzioni
del corpo interno dei cicli i,j, sono $n+(n-1)+(n-2)... = n(n+1)/2 = O(n^2/2)$ per ognuna di queste sottosequenze, si esegue la somma,
che nonostante sia una singola funzione, impiegherà tempo lineare nel numero degli elementi e quindi in generale un tempo $O(n)$,
per un totale complessivo di O(n^3).

Verifichiamolo sperimentalemente:


```python
def benchmark(v,solver):
    print("Benchmarking: ",solver.__name__)
    profiler = []
    N = 2600
    for n in range(200,N,200):
        start = timeit.default_timer()
        solver(v[0:n])
        end = timeit.default_timer()
        print(f"time for {n} is {end-start}")
        profiler.append(end-start)
    return list(range(200,N,200)),profiler
        
```

```python
n,t = benchmark(l,solve_one)
```

```python
plt.plot(n,t,label='measured');
n2 = [x**2/20**4 for x in n]
n3 = [x**3/20**6 for x in n]
plt.plot(n,n2,label='$n^2$');
plt.plot(n[:-2],n3[:-2],label='$n^3$');
plt.legend();
```

```python
#secondo tentativo di soluzione, miglioriamo il ciclo interno.

def solve_two(v):
    """ solve the maximum sum subsequence problem, solution 2 """
    # iterate all possibile subsequence with B > A
    # find the maximum
    max_sum = -10000000
    A,B = 0,0                # solutions
    n = len(v)
    for i in range(n):
        s = 0
        for j in range(i,n):
            s += v[j]
            if s > max_sum:
                max_sum = s
                A,B = i,j
    return (A,B,max_sum)
```

```python
# terzo algoritmo: buon algoritmica, linear time.
def solve_three(v):
    """ solve the maximum sum subsequence problem, solution 3 """
    # iterate just one time, find the maximum
    max_sum = -10000000
    s = 0, 
    b = 0       # solutions
    n = len(v)
    for i in range(n):
        s += v[i]
        if s > max_sum:
            max_sum = s
            A,B = b,i
        if s < 0:
            s = 0
            b = i+1            
    return (A,B,max_sum) 
```
