```python
import math

# Per ogni corpo:
# - posizione
# - velocità
# - massa
# - accelerazione

# Il sole sta in (0,0,0) con massa M1 >>>> m

# condizioni iniziali

r = [1, 0,   0 ]
v = [0, 0.5, 0]
a = [0, 0, 0]


# M1 + M2 = 1
dt = 0.0001
T = 10
N = int(T / dt)
print("# Precisione Integrazione (dt) = ",dt)
print("# Numero passi di Integrazione = ",N)
print("# Integro per ",T," passi.")

t = 0

print("{:.5f},{:.05f},{:.5f}".format(t,r[0],r[1]))

ekin_start = 0.5*v[0]*v[0]+v[1]*v[1]+v[2]*v[2]
epot_start = -1 / math.sqrt(r[0]*r[0] + r[1]*r[1] + r[2]*r[2])

while t <= T:
    r2 = r[0]*r[0] + r[1]*r[1] + r[2]*r[2]
    d = r2 * math.sqrt(r2)

    # nuova accelerazione
    a[0] = -r[0]/d
    a[1] = -r[1]/d
    a[2] = -r[2]/d

    # nuova velocità

    v[0] +=  a[0]*dt
    v[1] +=  a[1]*dt
    v[2] +=  a[2]*dt

    # nuova posizione
    r[0] +=  v[0]*dt
    r[1] +=  v[1]*dt
    r[2] +=  v[2]*dt

    print("{:.5f},{:.05f},{:.5f}".format(t,r[0],r[1]))
    t += dt

epot_finale = -1 / math.sqrt(r[0]*r[0] + r[1]*r[1] + r[2]*r[2])
ekin_finale = 0.5*v[0]*v[0]+v[1]*v[1]+v[2]*v[2]
print("Energia Ci. Iniziale = ",epot_start+ekin_start)
print("Energia Ci. Finale = ",epot_finale+ekin_finale)


```
